```python
%matplotlib notebook

import numpy as np
from matplotlib import pyplot as plt
from scipy.constants import hbar, k

from pytopo.mplplots.init_nb_plotting import *
```

```python
twopi = np.pi * 2
```

```python
def snr(t):
    return ((kappa * nbar * hbar * omega * t) / (k * TN))**.5 * np.sin(phi/2.)

kappa = twopi * 1e6
nbar = 5
TN = 0.2
phi = np.pi / 4.
omega = twopi * 5e9

tvals = np.linspace(0, 1e-3, 101)
snrvals = snr(tvals)

fig, ax = plt.subplots(1,1)
ax.plot(tvals * 1e3, snrvals)
ax.set_xlabel('Integration time (ms)')
ax.set_ylabel('SNR')
ax.grid(dashes = [1,1])
```
